```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('../Data/Frozen_Dessert_Production.csv',index_col='DATE',parse_dates=True)
```

```python
df
```

```python
df.columns = ['Production']
```

```python
df
```

```python
df.plot(figsize=(12,8))
```

```python
len(df)
```

```python
len(df)- 24
```

```python
test_size = 24
```

```python
test_ind = len(df)- test_size
```

```python
train = df.iloc[:test_ind]
test = df.iloc[test_ind:]
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(train)
```

```python
scaled_train = scaler.transform(train)
scaled_test = scaler.transform(test)
```

```python
#Time series Generator
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
```

```python
length = 18 
generator = TimeseriesGenerator(scaled_train,scaled_train,length=length
                               ,batch_size=1)
```

```python
X,y = generator[0]
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM
```

```python
n_feat = 1
```

```python
model = Sequential()
model.add(LSTM(100,input_shape=(length,n_feat)))
model.add(Dense(1))
model.compile(optimizer = 'adam',loss='mse')
```

```python
model.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop = EarlyStopping(monitor='val_loss',patience=2)
validation_generator = TimeseriesGenerator(scaled_test,scaled_test,
                                          length=length,batch_size=1)
```

```python
model.fit_generator(generator,epochs=20,validation_data=validation_generator
                   ,callbacks=[early_stop])
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses.plot()
```

```python
test_predictions = []

first_eval_batch = scaled_train[-length:]
current_batch = first_eval_batch.reshape((1,length,n_feat))

for i in range(len(test)):
    
    current_pred = model.predict(current_batch)[0]
    
    test_predictions.append(current_pred)
    
    current_batch = np.append(current_batch[:,1:,:],[[current_pred]],axis=1)
```

```python
#Inverse Transformation and comapre

true_predictions = scaler.inverse_transform(test_predictions)
```

```python
test['Predictions'] = true_predictions
```

```python
test.plot(figsize=(12,8))
```
